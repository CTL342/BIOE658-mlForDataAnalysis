```python
# Libraries
import random
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

random.seed(321)
```

```python
# Dataset
df = pd.read_csv('medical_data.csv')
print(df.head())
```

```python
# Results Dataframe
df_metrics = pd.DataFrame({}, index=['Log Reg', 'Ran For', 'SVC', 'KNN'])

# Data Preparation
X, y = df[['Age', 'Blood_Pressure', 'Cholesterol', 'Glucose']], df['Diagnosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# Implementation
print("Confusion Matrices")
# Logistic Regression
model_log = LogisticRegression().fit(X_train, y_train)
model_log_pred = model_log.predict(X_test)
print("Logistic Regression")
print(confusion_matrix(y_test, model_log_pred))
df_metrics.at['Log Reg', 'Accuracy'] = accuracy_score(y_test, model_log_pred)
df_metrics.at['Log Reg', 'Precision'] = precision_score(y_test, model_log_pred)
df_metrics.at['Log Reg', 'Recall'] = recall_score(y_test, model_log_pred)
df_metrics.at['Log Reg', 'F1 Score'] = f1_score(y_test, model_log_pred)
# Random Forest
model_rf = RandomForestClassifier().fit(X_train, y_train)
model_rf_pred = model_rf.predict(X_test)
print("Random Forest")
print(confusion_matrix(y_test, model_rf_pred))
df_metrics.at['Ran For', 'Accuracy'] = accuracy_score(y_test, model_rf_pred)
df_metrics.at['Ran For', 'Precision'] = precision_score(y_test, model_rf_pred)
df_metrics.at['Ran For', 'Recall'] = recall_score(y_test, model_rf_pred)
df_metrics.at['Ran For', 'F1 Score'] = f1_score(y_test, model_rf_pred)
# SVC
model_svc = SVC().fit(X_train, y_train)
model_svc_pred = model_svc.predict(X_test)
print("SVC")
print(confusion_matrix(y_test, model_svc_pred))
df_metrics.at['SVC', 'Accuracy'] = accuracy_score(y_test, model_svc_pred)
df_metrics.at['SVC', 'Precision'] = precision_score(y_test, model_svc_pred)
df_metrics.at['SVC', 'Recall'] = recall_score(y_test, model_svc_pred)
df_metrics.at['SVC', 'F1 Score'] = f1_score(y_test, model_svc_pred)
# KNN
model_knn = KNeighborsClassifier().fit(X_train, y_train)
model_knn_pred = model_knn.predict(X_test)
print("KNN")
print(confusion_matrix(y_test, model_knn_pred))
df_metrics.at['KNN', 'Accuracy'] = accuracy_score(y_test, model_knn_pred)
df_metrics.at['KNN', 'Precision'] = precision_score(y_test, model_knn_pred)
df_metrics.at['KNN', 'Recall'] = recall_score(y_test, model_knn_pred)
df_metrics.at['KNN', 'F1 Score'] = f1_score(y_test, model_knn_pred)

# Print Results
print("\nOther Metrics")
print(df_metrics)
```

```python
print('1. Random Forest performs best because it achieved the highest accuracy and precision, along with \none of the higher scores for recall and f1.')
print('2. The recall metric should be used to ensure you can identify as many actuall positive cases as possible.')
print('3. We should focus on f1 because it calculates the harmonic mean of precision and recall and penalizes \nthe model if the metric drops too low.')
print('4. Overfitting means the model is memorizing the training data and fail with unseen testing data while \nunderfitting makes a model unable to learn the relationship between features and output.')
print('5. Random forest is more likely to overfit because its decision trees can develop perfectly memorized training \npoints to create predicitons with no restraint.')
```
