```python
# Libraries
import random
import sklearn
import pandas as pd

random.seed(321)
```

```python
print("Problem 2")

df = sklearn.datasets.load_diabetes(as_frame=True)
print(df.data.head(2))
print(df.target.head(2))

df.target = (df.target > df.target.median()).astype(int)

X = df.data
y = df.target

X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.2, random_state=321)
```

```python
lr_model = sklearn.linear_model.LogisticRegression().fit(X_train, y_train)
print("Logistic Regression Summary")
print("Coefficients:", lr_model.coef_)
print("Intercept:", lr_model.intercept_)
print("Accuracy:", sklearn.metrics.accuracy_score(y_test, lr_model.predict(X_test)))
print("Precision:", sklearn.metrics.precision_score(y_test, lr_model.predict(X_test), average='weighted'))
print("Recall:", sklearn.metrics.recall_score(y_test, lr_model.predict(X_test), average='weighted'))
print("F1-Score:", sklearn.metrics.f1_score(y_test, lr_model.predict(X_test), average='weighted'))
```

```python
lr_model = sklearn.linear_model.LogisticRegression().fit(X_train, y_train)
print("Logistic Regression Summary")
print("Coefficients:", lr_model.coef_)
print("Intercept:", lr_model.intercept_)
print("Accuracy:", sklearn.metrics.accuracy_score(y_test, (lr_model.predict_proba(X_test)[:, 1] >= 0.3).astype(int)))
print("Precision:", sklearn.metrics.precision_score(y_test, (lr_model.predict_proba(X_test)[:, 1] >= 0.3).astype(int), average='weighted', zero_division=0))
print("Recall:", sklearn.metrics.recall_score(y_test, (lr_model.predict_proba(X_test)[:, 1] >= 0.3).astype(int), average='weighted'))
print("F1-Score:", sklearn.metrics.f1_score(y_test, (lr_model.predict_proba(X_test)[:, 1] >= 0.3).astype(int), average='weighted'))
```

```python
lr_model = sklearn.linear_model.LogisticRegression().fit(X_train, y_train)
print("Logistic Regression Summary")
print("Coefficients:", lr_model.coef_)
print("Intercept:", lr_model.intercept_)
print("Accuracy:", sklearn.metrics.accuracy_score(y_test, (lr_model.predict_proba(X_test)[:, 1] >= 0.7).astype(int)))
print("Precision:", sklearn.metrics.precision_score(y_test, (lr_model.predict_proba(X_test)[:, 1] >= 0.7).astype(int), average='weighted', zero_division=0))
print("Recall:", sklearn.metrics.recall_score(y_test, (lr_model.predict_proba(X_test)[:, 1] >= 0.7).astype(int), average='weighted'))
print("F1-Score:", sklearn.metrics.f1_score(y_test, (lr_model.predict_proba(X_test)[:, 1] >= 0.7).astype(int), average='weighted'))
```
