```python
# Libraries
import random
import sklearn
import pandas as pd

random.seed(321)
```

```python
print("Problem 3")

df = sklearn.datasets.load_diabetes(as_frame=True)
print(df.data.head(2))
print(df.target.head(2))

df.target = (df.target > df.target.median()).astype(int)

X = df.data
y = df.target

X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.2, random_state=321)
```

```python
knn_model1 = sklearn.neighbors.KNeighborsClassifier(n_neighbors=2).fit(X_train, y_train)
print("K-Nearest Neighbors (n_neighbors:2) Summary")
print("Accuracy:", sklearn.metrics.accuracy_score(y_test, knn_model1.predict(X_test)))
print("Precision:", sklearn.metrics.precision_score(y_test, knn_model1.predict(X_test), average='weighted'))
print("Recall:", sklearn.metrics.recall_score(y_test, knn_model1.predict(X_test), average='weighted'))
print("F1-Score:", sklearn.metrics.f1_score(y_test, knn_model1.predict(X_test), average='weighted'))
```

```python
knn_model2 = sklearn.neighbors.KNeighborsClassifier(n_neighbors=20).fit(X_train, y_train)
print("K-Nearest Neighbors (n_neighbors:20) Summary")
print("Accuracy:", sklearn.metrics.accuracy_score(y_test, knn_model2.predict(X_test)))
print("Precision:", sklearn.metrics.precision_score(y_test, knn_model2.predict(X_test), average='weighted'))
print("Recall:", sklearn.metrics.recall_score(y_test, knn_model2.predict(X_test), average='weighted'))
print("F1-Score:", sklearn.metrics.f1_score(y_test, knn_model2.predict(X_test), average='weighted'))
```

```python
knn_model3 = sklearn.neighbors.KNeighborsClassifier(n_neighbors=100).fit(X_train, y_train)
print("K-Nearest Neighbors (n_neighbors:100) Summary")
print("Accuracy:", sklearn.metrics.accuracy_score(y_test, knn_model3.predict(X_test)))
print("Precision:", sklearn.metrics.precision_score(y_test, knn_model3.predict(X_test), average='weighted'))
print("Recall:", sklearn.metrics.recall_score(y_test, knn_model3.predict(X_test), average='weighted'))
print("F1-Score:", sklearn.metrics.f1_score(y_test, knn_model3.predict(X_test), average='weighted'))
```
