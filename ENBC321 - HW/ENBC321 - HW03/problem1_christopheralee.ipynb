```python
# Libraries
import random
import pandas as pd
import sklearn
from mlxtend.feature_selection import SequentialFeatureSelector 

# Seed
random.seed(321)
```

```python
# Dataset Link: https://www.kaggle.com/datasets/dileep070/heart-disease-prediction-using-logistic-regression
print("Problem 1")
df = pd.read_csv('framingham.csv').dropna()
print(df.head())
X = df[['male', 'age', 'education', 'currentSmoker', 'cigsPerDay', 'BPMeds', 'prevalentStroke', 'prevalentHyp', 'diabetes']]
y = df['totChol']
```

```python
print("Part 1")
all_features = X.columns.tolist()
model_all = sklearn.linear_model.LinearRegression().fit(X, y)
model_all_r2 = model_all.score(X, y)
model_all_mse = sklearn.metrics.mean_squared_error(y, model_all.predict(X))

print("All In or Full Model Feature Selection")
print("Features:", all_features)
print(f"R^2 Value: {model_all_r2:.6f}")
print(f"MSE Value: {model_all_mse:.6f}")
```

```python
print("Part 2")
rfe = sklearn.feature_selection.RFE(estimator=sklearn.linear_model.LinearRegression(), n_features_to_select=4).fit(X, y)
back_elim_features = X.columns[rfe.support_].tolist()

model_back_elim = sklearn.linear_model.LinearRegression().fit(X[back_elim_features], y)
model_back_elim_r2 = model_back_elim.score(X[back_elim_features], y)
model_back_elim_mse = sklearn.metrics.mean_squared_error(y, model_back_elim.predict(X[back_elim_features]))

print("Backwards Elimination Model Feature Selection")
print("Features:", back_elim_features)
print(f"R^2 Value: {model_back_elim_r2:.6f}")
print(f"MSE Value: {model_back_elim_mse:.6f}")
```

```python
print("Part 3")
sfs = sklearn.feature_selection.SequentialFeatureSelector(estimator=sklearn.linear_model.LinearRegression(), n_features_to_select=4, direction='forward').fit(X, y)
forw_sele_features = X.columns[sfs.get_support()].tolist()

model_forw_sele = sklearn.linear_model.LinearRegression().fit(X[forw_sele_features], y)
model_forw_sele_r2 = model_forw_sele.score(X[forw_sele_features], y)
model_forw_sele_mse = sklearn.metrics.mean_squared_error(y, model_forw_sele.predict(X[forw_sele_features]))

print("Forward Selection Model Feature Selection")
print("Features:", forw_sele_features)
print(f"R^2 Value: {model_forw_sele_r2:.6f}")
print(f"MSE Value: {model_forw_sele_mse:.6f}")
```

```python
print("Part 4")
sfs_bidi = SequentialFeatureSelector(estimator=sklearn.linear_model.LinearRegression(),
                                                               k_features=4,
                                                               forward=True,
                                                               floating=True,
                                                               scoring='r2',
                                                               cv=0)
sfs_bidi.fit(X, y)
bidi_features = list(sfs_bidi.k_feature_names_)
model_bidi = sklearn.linear_model.LinearRegression().fit(X[bidi_features], y)
model_bidi_elim_r2 = model_bidi.score(X[bidi_features], y)
model_bidi_elim_mse = sklearn.metrics.mean_squared_error(y, model_bidi.predict(X[bidi_features]))

print("Bidirectional Elimination Model Feature Selection")
print("Features:", bidi_features)
print(f"R^2 Value: {model_bidi_elim_r2:.6f}")
print(f"MSE Value: {model_bidi_elim_mse:.6f}")
```

```python
print("Part 5")
anova = sklearn.feature_selection.SelectKBest(score_func=sklearn.feature_selection.f_regression, k=4).fit(X, y)
anova_features = X.columns[anova.get_support()].tolist()

model_anova = sklearn.linear_model.LinearRegression().fit(X[anova_features], y)
model_anova_ftest_r2 = model_anova.score(X[anova_features], y)
model_anova_ftest_mse = sklearn.metrics.mean_squared_error(y, model_anova.predict(X[anova_features]))

print("Anova F-Test Model Feature Selection")
print("Features:", anova_features)
print(f"R^2 Value: {model_anova_ftest_r2:.6f}")
print(f"MSE Value: {model_anova_ftest_mse:.6f}")
```

```python
print("Part 6")
mi = sklearn.feature_selection.SelectKBest(score_func=sklearn.feature_selection.mutual_info_regression, k=4).fit(X, y)
mi_features = X.columns[mi.get_support()].tolist()

model_mi = sklearn.linear_model.LinearRegression().fit(X[mi_features], y)
model_mi_r2 = model_mi.score(X[mi_features], y)
model_mi_mse = sklearn.metrics.mean_squared_error(y, model_mi.predict(X[mi_features]))

print("Mutual Information Model Feature Selection")
print("Features:", mi_features)
print(f"R^2 Value: {model_mi_r2:.6f}")
print(f"MSE Value: {model_mi_mse:.6f}")
```

```python
print("Part 7")
correlations = X.apply(lambda col: col.corr(y)).abs()
pearson_features = correlations.nlargest(4).index.tolist()

model_pearson = sklearn.linear_model.LinearRegression().fit(X[pearson_features], y)
model_pearson_r2 = model_pearson.score(X[pearson_features], y)
model_pearson_mse = sklearn.metrics.mean_squared_error(y, model_pearson.predict(X[pearson_features]))

print("Pearson Correlation Model Feature Selection")
print("Features:", pearson_features)
print(f"R^2 Value: {model_pearson_r2:.6f}")
print(f"MSE Value: {model_pearson_mse:.6f}")
```

```python
print("Final Feature Selection Ranking")
print("1. Bidirectional Elimination")
print("2. Forward Selection")
print("3. ANOVA F-Test")
print("4. Pearson Correlation")
print("5. Mutual Information")
print("6. Backward Elimination")
print("7. All-in or Full Model\n")

print("The following ranking is based on R-squared value (greatest to least), and MSE (least to greatest)")
print("An exception to this trend does occur where all-in or full model feature selection is put last due \nfailing to drop any features which was the whole point of feature selection, even if it \ntechnically yields the highest r-squared value.")
```
