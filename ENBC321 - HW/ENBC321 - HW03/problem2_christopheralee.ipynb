```python
# Libraries
import random
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import numpy as np

# Seed
random.seed(321)
```

```python
print("Problem 2")

df = pd.read_csv('gene_expression_data.csv')
X = df['Gene_Expression'].values.reshape(-1, 1)
y = df['Protein_Expression'].values

X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.2, random_state=321)

model_svm = sklearn.svm.SVR(kernel='rbf', C=100, epsilon=0.1).fit(X_train, y_train)
y_pred = model_svm.predict(X_test)

plt.scatter(y_test, y_pred)
plt.title("True Protein Expresssion vs. Predicted Protein Expression")
plt.xlabel("True Protein Expression")
plt.ylabel("Predicted Protein Expression")
plt.show()
```

```python
# Linear and nonlinear kernels visualization plots
X_curve = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
fig, axes = plt.subplots(2, 2, figsize=(10, 6))
fig.suptitle("SVR Regression: Linear vs. RBF at Epsilon 0.1 and 4")
plt.subplots_adjust(hspace=0.4)

# Plot 1
model_lin1 = sklearn.svm.SVR(kernel='linear', C=100, epsilon=0.1).fit(X_train, y_train)
y_curve_lin1 = model_lin1.predict(X_curve)
y_pred_lin1 = model_lin1.predict(X_test)
r2_lin1 = sklearn.metrics.r2_score(y_test, y_pred_lin1)
mse_lin1 = sklearn.metrics.mean_squared_error(y_test, y_pred_lin1)

axes[0, 0].scatter(X_test, y_test, label='Actual Data')
axes[0, 0].plot(X_curve, y_curve_lin1, color='red', linewidth=2, label='SVR Curve')
axes[0, 0].set_title(f"Plot 1\nKernel: Linear | Epsilon: 0.1 | R-Squared: {r2_lin1:.2f} | MSE: {mse_lin1:.2f}", fontsize=7)
axes[0, 0].set_xlabel("Gene Expression", fontsize=7)
axes[0, 0].set_ylabel("Protein Expression", fontsize=7)
axes[0, 0].grid()
axes[0, 0].legend()

# Plot 2
model_lin2 = sklearn.svm.SVR(kernel='linear', C=100, epsilon=4).fit(X_train, y_train)
y_curve_lin2 = model_lin2.predict(X_curve)
y_pred_lin2 = model_lin2.predict(X_test)
r2_lin2 = sklearn.metrics.r2_score(y_test, y_pred_lin2)
mse_lin2 = sklearn.metrics.mean_squared_error(y_test, y_pred_lin2)

axes[0, 1].scatter(X_test, y_test, label='Actual Data')
axes[0, 1].plot(X_curve, y_curve_lin2, color='red', linewidth=2, label='SVR Curve')
axes[0, 1].set_title(f"Plot 2\nKernel: Linear | Epsilon: 4 | R-Squared: {r2_lin2:.2f} | MSE: {mse_lin2:.2f}", fontsize=7)
axes[0, 1].set_xlabel("Gene Expression", fontsize=7)
axes[0, 1].set_ylabel("Protein Expression", fontsize=7)
axes[0, 1].grid()
axes[0, 1].legend()

# Plot 3
model_lin3 = sklearn.svm.SVR(kernel='rbf', C=100, epsilon=0.1).fit(X_train, y_train)
y_curve_lin3 = model_lin3.predict(X_curve)
y_pred_lin3 = model_lin3.predict(X_test)
r2_lin3 = sklearn.metrics.r2_score(y_test, y_pred_lin3)
mse_lin3 = sklearn.metrics.mean_squared_error(y_test, y_pred_lin3)

axes[1, 0].scatter(X_test, y_test, label='Actual Data')
axes[1, 0].plot(X_curve, y_curve_lin3, color='red', linewidth=2, label='SVR Curve')
axes[1, 0].set_title(f"Plot 3\nKernel: rbf | Epsilon: 0.1 | R-Squared: {r2_lin3:.2f} | MSE: {mse_lin3:.2f}", fontsize=7)
axes[1, 0].set_xlabel("Gene Expression", fontsize=7)
axes[1, 0].set_ylabel("Protein Expression", fontsize=7)
axes[1, 0].grid()
axes[1, 0].legend()

# Plot 4
model_lin4 = sklearn.svm.SVR(kernel='rbf', C=100, epsilon=4).fit(X_train, y_train)
y_curve_lin4 = model_lin4.predict(X_curve)
y_pred_lin4 = model_lin4.predict(X_test)
r2_lin4 = sklearn.metrics.r2_score(y_test, y_pred_lin4)
mse_lin4 = sklearn.metrics.mean_squared_error(y_test, y_pred_lin4)

axes[1, 1].scatter(X_test, y_test, label='Actual Data')
axes[1, 1].plot(X_curve, y_curve_lin4, color='red', linewidth=2, label='SVR Curve')
axes[1, 1].set_title(f"_\n\nPlot 4\nKernel: rbf | Epsilon: 4 | R-Squared: {r2_lin4:.2f} | MSE: {mse_lin4:.2f}", fontsize=7)
axes[1, 1].set_xlabel("Gene Expression", fontsize=7)
axes[1, 1].set_ylabel("Protein Expression", fontsize=7)
axes[1, 1].grid()
axes[1, 1].legend()

plt.show()
```

```python
print("In Support Vector Regression, a small epsilon value enforces a more strict margin of error and causes the Linear kernel to struggle \nwith finding the most optimal trend and fit the data points, unlike the RBF kernel which was more flexible to fit individual data points \nyielding a high accuracy, but risks overfitting. A large epsilon value of 4 creates a less strict margin where most errors are not \nactually penalized. However, the negative effects include significantly worse regression curbes and avoidance data which leads to \ndrastic underfitting, causing the Mean Squared Error to spike and overall prediction accruacy to drop.")
```
