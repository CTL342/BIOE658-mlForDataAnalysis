```python
# Libraries
import random
import sklearn
import pandas as pd
from matplotlib import pyplot as plt

random.seed(321)
```

```python
# Import data
df = pd.read_csv('gene_expression_data.csv')
print(df.head())
```

```python
print("Problem 1")

# Prepare Data
X, y = df[['Gene_Expression']], df['Protein_Expression']
pro_exp_med = y.median()

y[y > pro_exp_med] = 1
y[y != 1] = 0

print(y.head())

# Model Training & Testing
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.2, random_state=321)
# Linear Model
model_svm1 = sklearn.svm.SVC(kernel='linear').fit(X_train, y_train)
pred1 = model_svm1.predict(X_test)
# RBF Model, C = 0.01
model_svm2 = sklearn.svm.SVC(kernel='rbf', C=0.01).fit(X_train, y_train)
pred2 = model_svm2.predict(X_test)
# RBF Model, C = 0.1
model_svm3 = sklearn.svm.SVC(kernel='rbf', C=0.1).fit(X_train, y_train)
pred3 = model_svm3.predict(X_test)
# RBF Model, C = 1
model_svm4 = sklearn.svm.SVC(kernel='rbf', C=1).fit(X_train, y_train)
pred4 = model_svm4.predict(X_test)
# RBF Model, C = 10
model_svm5 = sklearn.svm.SVC(kernel='rbf', C=10).fit(X_train, y_train)
pred5 = model_svm5.predict(X_test)
# RBF Model, C = 100
model_svm6 = sklearn.svm.SVC(kernel='rbf', C=100).fit(X_train, y_train)
pred6 = model_svm6.predict(X_test)


# Plot Predictions
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
fig, axes = plt.subplots(2, 3, figsize=(14, 7))
fig.suptitle("SVM Classification: Linear & RBF with Different C Values")
plt.subplots_adjust(hspace=0.4)


axes[0, 0].scatter(X_test, y_test, color='blue', label='Actual')
axes[0, 0].scatter(X_test, pred1, color='red', marker='x', label='Predicted')
axes[0, 0].set_title(f"Plot 1 | Kernel: Linear\nAccuracy: {accuracy_score(y_test, pred1):.4f} | Precision: {precision_score(y_test, pred1):.4f} | Recall: {recall_score(y_test, pred1):.4f} | F1-Score: {f1_score(y_test, pred1):.4f}", fontsize=7)
axes[0, 0].set_xlabel("Gene Expression", fontsize=7)
axes[0, 0].set_ylabel("Protein Expression", fontsize=7)
axes[0, 0].grid()
axes[0, 0].legend()

axes[0, 1].scatter(X_test, y_test, color='blue', label='Actual')
axes[0, 1].scatter(X_test, pred2, color='red', marker='x', label='Predicted')
axes[0, 1].set_title(f"Plot 2 | Kernel: RBF | C: 0.01\nAccuracy: {accuracy_score(y_test, pred2):.4f} | Precision: {precision_score(y_test, pred2):.4f} | Recall: {recall_score(y_test, pred2):.4f} | F1-Score: {f1_score(y_test, pred2):.4f}", fontsize=7)
axes[0, 1].set_xlabel("Gene Expression", fontsize=7)
axes[0, 1].set_ylabel("Protein Expression", fontsize=7)
axes[0, 1].grid()
axes[0, 1].legend()

axes[0, 2].scatter(X_test, y_test, color='blue', label='Actual')
axes[0, 2].scatter(X_test, pred3, color='red', marker='x', label='Predicted')
axes[0, 2].set_title(f"Plot 3 | Kernel: RBF | C: 0.1\nAccuracy: {accuracy_score(y_test, pred3):.4f} | Precision: {precision_score(y_test, pred3):.4f} | Recall: {recall_score(y_test, pred3):.4f} | F1-Score: {f1_score(y_test, pred3):.4f}", fontsize=7)
axes[0, 2].set_xlabel("Gene Expression", fontsize=7)
axes[0, 2].set_ylabel("Protein Expression", fontsize=7)
axes[0, 2].grid()
axes[0, 2].legend()

axes[1, 0].scatter(X_test, y_test, color='blue', label='Actual')
axes[1, 0].scatter(X_test, pred4, color='red', marker='x', label='Predicted')
axes[1, 0].set_title(f"Plot 4 | Kernel: RBF | C: 1\nAccuracy: {accuracy_score(y_test, pred4):.4f} | Precision: {precision_score(y_test, pred4):.4f} | Recall: {recall_score(y_test, pred4):.4f} | F1-Score: {f1_score(y_test, pred4):.4f}", fontsize=7)
axes[1, 0].set_xlabel("Gene Expression", fontsize=7)
axes[1, 0].set_ylabel("Protein Expression", fontsize=7)
axes[1, 0].grid()
axes[1, 0].legend()

axes[1, 1].scatter(X_test, y_test, color='blue', label='Actual')
axes[1, 1].scatter(X_test, pred5, color='red', marker='x', label='Predicted')
axes[1, 1].set_title(f"Plot 5 | Kernel: RBF | C: 10\nAccuracy: {accuracy_score(y_test, pred5):.4f} | Precision: {precision_score(y_test, pred5):.4f} | Recall: {recall_score(y_test, pred5):.4f} | F1-Score: {f1_score(y_test, pred5):.4f}", fontsize=7)
axes[1, 1].set_xlabel("Gene Expression", fontsize=7)
axes[1, 1].set_ylabel("Protein Expression", fontsize=7)
axes[1, 1].grid()
axes[1, 1].legend()

axes[1, 2].scatter(X_test, y_test, color='blue', label='Actual')
axes[1, 2].scatter(X_test, pred6, color='red', marker='x', label='Predicted')
axes[1, 2].set_title(f"Plot 6 | Kernel: RBF | C: 100\nAccuracy: {accuracy_score(y_test, pred6):.4f} | Precision: {precision_score(y_test, pred6):.4f} | Recall: {recall_score(y_test, pred6):.4f} | F1-Score: {f1_score(y_test, pred6):.4f}", fontsize=7)
axes[1, 2].set_xlabel("Gene Expression", fontsize=7)
axes[1, 2].set_ylabel("Protein Expression", fontsize=7)
axes[1, 2].grid()
axes[1, 2].legend()
```
