```python
# Libraries
import random
import sklearn
import pandas as pd
from matplotlib import pyplot as plt

random.seed(321)
```

```python
# Import data
X = pd.DataFrame(sklearn.datasets.load_breast_cancer()['data'])
y = pd.Series(sklearn.datasets.load_breast_cancer()['target'])
print("Features\n",X.head())
print("\nTarget\n", y.head())
```

```python
print("Problem 3")
# Datatable to store values
df_dt = pd.DataFrame({}, index=['md:3,  mss:2', 'md:3,  mss:5', 'md:3,  mss:10', 'md:5,  mss:2', 'md:5,  mss:5', 'md:5,  mss:10', 'md:10, mss:2', 'md:10, mss:5', 'md:10, mss:10'])

df_rf = pd.DataFrame({}, index=['ne:10', 'ne:50', 'ne:100', 'ne:200'])

# Model Training & Testing
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.2, random_state=123)

# Decision Tree Classifier Model
from sklearn.tree import DecisionTreeClassifier
# Decision Tree, Max Depth 3, Min Sample Split 2
model_dt1 = DecisionTreeClassifier(max_depth=3, min_samples_split=2).fit(X_train, y_train)
model_dt1_pred = model_dt1.predict(X_test)
df_dt.at['md:3,  mss:2', 'Accuracy'] = accuracy_score(y_test, model_dt1_pred)
df_dt.at['md:3,  mss:2', 'Precision'] = precision_score(y_test, model_dt1_pred)
df_dt.at['md:3,  mss:2', 'Recall'] = recall_score(y_test, model_dt1_pred)
df_dt.at['md:3,  mss:2', 'F1-Score'] = f1_score(y_test, model_dt1_pred)
# Decision Tree, Max Depth 3, Min Sample Split 5
model_dt2 = DecisionTreeClassifier(max_depth=3, min_samples_split=5).fit(X_train, y_train)
model_dt2_pred = model_dt2.predict(X_test)
df_dt.at['md:3,  mss:5', 'Accuracy'] = accuracy_score(y_test, model_dt2_pred)
df_dt.at['md:3,  mss:5', 'Precision'] = precision_score(y_test, model_dt2_pred)
df_dt.at['md:3,  mss:5', 'Recall'] = recall_score(y_test, model_dt2_pred)
df_dt.at['md:3,  mss:5', 'F1-Score'] = f1_score(y_test, model_dt2_pred)
# Decision Tree, Max Depth 3, Min Sample Split 10
model_dt3 = DecisionTreeClassifier(max_depth=3, min_samples_split=10).fit(X_train, y_train)
model_dt3_pred = model_dt3.predict(X_test)
df_dt.at['md:3,  mss:10', 'Accuracy'] = accuracy_score(y_test, model_dt3_pred)
df_dt.at['md:3,  mss:10', 'Precision'] = precision_score(y_test, model_dt3_pred)
df_dt.at['md:3,  mss:10', 'Recall'] = recall_score(y_test, model_dt3_pred)
df_dt.at['md:3,  mss:10', 'F1-Score'] = f1_score(y_test, model_dt3_pred)
# Decision Tree, Max Depth 5, Min Sample Split 2
model_dt4 = DecisionTreeClassifier(max_depth=5, min_samples_split=2).fit(X_train, y_train)
model_dt4_pred = model_dt4.predict(X_test)
df_dt.at['md:5,  mss:2', 'Accuracy'] = accuracy_score(y_test, model_dt4_pred)
df_dt.at['md:5,  mss:2', 'Precision'] = precision_score(y_test, model_dt4_pred)
df_dt.at['md:5,  mss:2', 'Recall'] = recall_score(y_test, model_dt4_pred)
df_dt.at['md:5,  mss:2', 'F1-Score'] = f1_score(y_test, model_dt4_pred)
# Decision Tree, Max Depth 5, Min Sample Split 5
model_dt5 = DecisionTreeClassifier(max_depth=5, min_samples_split=5).fit(X_train, y_train)
model_dt5_pred = model_dt5.predict(X_test)
df_dt.at['md:5,  mss:5', 'Accuracy'] = accuracy_score(y_test, model_dt5_pred)
df_dt.at['md:5,  mss:5', 'Precision'] = precision_score(y_test, model_dt5_pred)
df_dt.at['md:5,  mss:5', 'Recall'] = recall_score(y_test, model_dt5_pred)
df_dt.at['md:5,  mss:5', 'F1-Score'] = f1_score(y_test, model_dt5_pred)
# Decision Tree, Max Depth 5, Min Sample Split 10
model_dt6 = DecisionTreeClassifier(max_depth=5, min_samples_split=10).fit(X_train, y_train)
model_dt6_pred = model_dt6.predict(X_test)
df_dt.at['md:5,  mss:10', 'Accuracy'] = accuracy_score(y_test, model_dt6_pred)
df_dt.at['md:5,  mss:10', 'Precision'] = precision_score(y_test, model_dt6_pred)
df_dt.at['md:5,  mss:10', 'Recall'] = recall_score(y_test, model_dt6_pred)
df_dt.at['md:5,  mss:10', 'F1-Score'] = f1_score(y_test, model_dt6_pred)
# Decision Tree, Max Depth 10, Min Sample Split 2
model_dt7 = DecisionTreeClassifier(max_depth=10, min_samples_split=2).fit(X_train, y_train)
model_dt7_pred = model_dt7.predict(X_test)
df_dt.at['md:10, mss:2', 'Accuracy'] = accuracy_score(y_test, model_dt7_pred)
df_dt.at['md:10, mss:2', 'Precision'] = precision_score(y_test, model_dt7_pred)
df_dt.at['md:10, mss:2', 'Recall'] = recall_score(y_test, model_dt7_pred)
df_dt.at['md:10, mss:2', 'F1-Score'] = f1_score(y_test, model_dt7_pred)
# Decision Tree, Max Depth 10, Min Sample Split 5
model_dt8 = DecisionTreeClassifier(max_depth=10, min_samples_split=5).fit(X_train, y_train)
model_dt8_pred = model_dt8.predict(X_test)
df_dt.at['md:10, mss:5', 'Accuracy'] = accuracy_score(y_test, model_dt8_pred)
df_dt.at['md:10, mss:5', 'Precision'] = precision_score(y_test, model_dt8_pred)
df_dt.at['md:10, mss:5', 'Recall'] = recall_score(y_test, model_dt8_pred)
df_dt.at['md:10, mss:5', 'F1-Score'] = f1_score(y_test, model_dt8_pred)
# Decision Tree, Max Depth 10, Min Sample Split 10
model_dt9 = DecisionTreeClassifier(max_depth=10, min_samples_split=10).fit(X_train, y_train)
model_dt9_pred = model_dt9.predict(X_test)
df_dt.at['md:10, mss:10', 'Accuracy'] = accuracy_score(y_test, model_dt9_pred)
df_dt.at['md:10, mss:10', 'Precision'] = precision_score(y_test, model_dt9_pred)
df_dt.at['md:10, mss:10', 'Recall'] = recall_score(y_test, model_dt9_pred)
df_dt.at['md:10, mss:10', 'F1-Score'] = f1_score(y_test, model_dt9_pred)

# Random Forest Classifier Model
from sklearn.ensemble import RandomForestClassifier
# Random Forest, 10 Trees
model_rf1 = RandomForestClassifier(n_estimators=10).fit(X_train, y_train)
model_rf1_pred = model_rf1.predict(X_test)
df_rf.at['ne:10', 'Accuracy'] = accuracy_score(y_test, model_rf1_pred)
df_rf.at['ne:10', 'Precision'] = precision_score(y_test, model_rf1_pred)
df_rf.at['ne:10', 'Recall'] = recall_score(y_test, model_rf1_pred)
df_rf.at['ne:10', 'F1-Score'] = f1_score(y_test, model_rf1_pred)
# Random Forest, 50 Trees
model_rf2 = RandomForestClassifier(n_estimators=50).fit(X_train, y_train)
model_rf2_pred = model_rf2.predict(X_test)
df_rf.at['ne:50', 'Accuracy'] = accuracy_score(y_test, model_rf2_pred)
df_rf.at['ne:50', 'Precision'] = precision_score(y_test, model_rf2_pred)
df_rf.at['ne:50', 'Recall'] = recall_score(y_test, model_rf2_pred)
df_rf.at['ne:50', 'F1-Score'] = f1_score(y_test, model_rf2_pred)
# Random Forest, 100 Trees
model_rf3 = RandomForestClassifier(n_estimators=100).fit(X_train, y_train)
model_rf3_pred = model_rf3.predict(X_test)
df_rf.at['ne:100', 'Accuracy'] = accuracy_score(y_test, model_rf3_pred)
df_rf.at['ne:100', 'Precision'] = precision_score(y_test, model_rf3_pred)
df_rf.at['ne:100', 'Recall'] = recall_score(y_test, model_rf3_pred)
df_rf.at['ne:100', 'F1-Score'] = f1_score(y_test, model_rf3_pred)
# Random Forest, 200 Trees
model_rf4 = RandomForestClassifier(n_estimators=200).fit(X_train, y_train)
model_rf4_pred = model_rf4.predict(X_test)
df_rf.at['ne:200', 'Accuracy'] = accuracy_score(y_test, model_rf4_pred)
df_rf.at['ne:200', 'Precision'] = precision_score(y_test, model_rf4_pred)
df_rf.at['ne:200', 'Recall'] = recall_score(y_test, model_rf4_pred)
df_rf.at['ne:200', 'F1-Score'] = f1_score(y_test, model_rf4_pred)

print("Legend Parameters")
print("md: max_depth")
print("mss: min_samples_split")
print("ne: n_estimators\n")

print("Decision Tree Results")
print(df_dt)

print("\nRandom Forest Results")
print(df_rf)

print("\nDecision Tree Analysis")
print("Increaseing the max_depth to 10 improved overall accuracy and precsiion, while adjusting the min_amples_split had little to no impact.")
print("\nRandom Forest Analysis")
print("Model achieved almost peak performance within 10 trees and convereged completely. Only small improvements when scaled to 200 trees.")
```

```python
print("1. A Decision Tree might overfit the training data because they can grow deep enough to create a specific \nrule for every single data point and can end up memorizing the data if that occurs.")
print("2. Random Forrests reduce overfitting because they train multiple independent trees on different subsets of \ndata and they use the bootstrapping method as well.")
print("3. Decision Trees usually have low bias but high variance while Random Forests also has a low bias but \nless variance than the Decision Trees.")
print("4. Increasing the n_estimators improves overall scores for accuracy, precision, recall, and f1. However, \nthe addition of more trees will eventually reach a point where it only increases the computational costs \nwithout reducing the error.")
```
