```python
# Import data
X = pd.DataFrame(sklearn.datasets.load_breast_cancer()['data'])
y = pd.DataFrame(sklearn.datasets.load_breast_cancer()['target'])
print("Features\n",X.head())
print("\nTarget\n", y.head())
```

```python
print("Problem 2")
# Model Training & Testing
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.2, random_state=321)
# Gaussian Naive Bayes (GNB) Classifier Model
model_gnb = sklearn.naive_bayes.GaussianNB().fit(X_train, y_train)
pred = model_gnb.predict(X_test)

# Model Evaluation
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
print("Accuracy:", accuracy_score(y_test, pred))
print("Precision:", precision_score(y_test, pred))
print("Recall:", recall_score(y_test, pred))
print("F1:", f1_score(y_test, pred))
print("\nClassification Report for Gaussian Naive Bayes (GNB) Classifier Model\n", classification_report(y_test, pred))

import seaborn as sns
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, pred)
print("Confusion Matrix\n", cm)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
print("1. Gaussian Naive Bayes assumes a normal distribution to simplify the calculation for the likelihood of each feature, specifcally using \nthe mean and variance only. When the features don't follow a normal distribution, the model will produce inaccurate probability estimates.")
print("2. The performance of the model would increase if a different classification algorithm such as Logistic Regression was used due to its \nnature of not being strictly reliant on the original assumption that the data follows a normal distribution. Gaussian Naive Bayes would only \nbe appropriate if we knew that the data itself followed a normal distribution pattern. Otherwise, choosing a different classification algorithm \nwould be ideal.")
```
